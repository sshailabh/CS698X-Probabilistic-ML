```python
# defining x and y 
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(11)
sigma2 = 0.05 # variance
x = np.linspace(0,4*np.pi,100) # x inputs
y = np.sin(x) + np.random.normal(loc=0.0, scale=np.sqrt(sigma2), size=x.shape[0]) # sine function with noise
print("Shape of input x and y",x.shape,y.shape) #check shape
```

```python
# define kernel matrix
def kernel_m(l,x): # kernel matrix formation
  var_mat = np.zeros((x.shape[0],x.shape[0]),dtype=float)
  for i in range(x.shape[0]):
    for j in range(x.shape[0]):
      var_mat[i][j] = np.exp((-1/(l**2)) *((x[i]-x[j])**2))
      if i==j:
        var_mat[i][j] += 1.0e-8 # adding small number to diagonal entry
  return var_mat
```

#l=0.2

```python
# Case l=0.2
l=0.2
K = kernel_m(l,x)
prior_samp = np.random.multivariate_normal(mean=np.zeros(x.shape[0]), cov=K) # prior random samples
mean_p = np.matmul(np.linalg.inv(np.linalg.inv(K) + np.eye(x.shape[0])*(1/sigma2)), y/sigma2) # mean of GP posterior
#plot
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, prior_samp, label='Prior samples') ## Plot prior samples
ax.plot(x, mean_p, label='Posterior mean') ## Plot mean posterior   
ax.plot(x, np.sin(x), label='True sine fxn') ## Plot true sine function
ax.set_xlabel('x')
ax.set_ylabel('function value')
ax.set_title('Plot for l=0.2')
leg = ax.legend(loc='best', frameon=True)
#print("Shape of prior and mean posterior",prior_samp.shape, mean_p.shape)
plt.savefig("question_4_0.2.png")
```
