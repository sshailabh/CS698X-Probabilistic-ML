# 6(a)

```python
def prepare_data(T,k):
    temp1 =  np.ones((T.shape[0],1))
    if k==1:
        temp2 = np.concatenate((temp1,T),axis=1)
    elif k==2:
        temp2 = np.concatenate((temp1,T,np.square(T)),axis=1)
    elif k==3:
        temp2 = np.concatenate((temp1,T,np.square(T),np.power(T,3)),axis=1)
    else:
        temp2 = np.concatenate((temp1,T,np.square(T),np.power(T,3),np.power(T,4)),axis=1)
    return temp2
```

```python
X = np.asarray([-2.23, -1.30, -0.42, 0.30, 0.33, 0.52, 0.87, 1.80, 2.74, 3.62], dtype=np.float32)[:,np.newaxis]
Y = np.asarray([1.01, 0.69, -0.66, -1.34, -1.75, -0.98, 0.25, 1.57, 1.65, 1.51], dtype=np.float32)[:,np.newaxis]
beta= 4
n_samp = 50 #100
X_samp = np.linspace(-4,4,n_samp)[:,np.newaxis]
```

```python
k=1 
## For k = 1
pdata_k1 = prepare_data(X,k) 
cov_mat_k1 = np.linalg.inv(np.matmul(pdata_k1.T,pdata_k1) + np.identity(k+1)) # covariance of posterior gaussian
mean_k1 = np.matmul(np.matmul(np.linalg.inv(np.matmul(pdata_k1.T,pdata_k1) + (np.identity(k+1)/beta)),pdata_k1.T),Y) # mean
w_a_k1 = np.random.multivariate_normal(np.squeeze(mean_k1,axis=1), cov_mat_k1, 10) # 10 random weights
y_samples_k1 = np.matmul(w_a_k1,prepare_data(X_samp,k).T) #predicted values of samples
fig, ax = plt.subplots()
lk1 = []
for i in range(y_samples_k1.shape[0]):
    lk1 += ax.plot(X_samp, y_samples_k1[i], color='blue')
lk1 += ax.plot(np.squeeze(X,axis=1),np.squeeze(Y,axis=1),'bo', color='red')
ax.set_ylim([-4,4])
ax.set_xlim([-4,4])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(lk1[-2:], ['Posterior', 'Original Data'], loc='best', frameon=True)
plt.savefig("question_6a_1.png")
```

```python
#Using MLE solution of W and marginal log lihood
#k=1
from scipy.stats import norm
from scipy.stats import multivariate_normal
k=1
wmle_k1 = np.squeeze(np.matmul(np.matmul(np.linalg.inv(np.matmul(pdata_k1.T,pdata_k1)),pdata_k1.T),Y)) #MLE
LMLE_k1 = 0
for i in range(Y.shape[0]):
    LMLE_k1 += np.log(norm.pdf(Y[i],loc=np.dot(wmle_k1,pdata_k1[i]),scale=np.sqrt(1/beta)) + multivariate_normal.pdf(wmle_k1) - multivariate_normal.pdf(wmle_k1,mean=np.squeeze(mean_k1),cov=cov_mat_k1))
print("MLE for K=1",LMLE_k1)
#print(np.log(norm.pdf(Y[0],loc=np.dot(np.squeeze(wmle_k1),pdata_k1[0]),scale = np.sqrt(1/beta))))
k=2 ###k=2
wmle_k2 = np.matmul(np.matmul(np.linalg.inv(np.matmul(pdata_k2.T,pdata_k2)),pdata_k2.T),Y)
LMLE_k2 = 0
for i in range(Y.shape[0]):
    LMLE_k2 += np.log(norm.pdf(Y[i],loc=np.dot(np.squeeze(wmle_k2),pdata_k2[i]),scale = np.sqrt(1/beta)))
print("MLE for K=2",LMLE_k2)
k=3 ###k=3
wmle_k3 = np.matmul(np.matmul(np.linalg.inv(np.matmul(pdata_k3.T,pdata_k3)),pdata_k3.T),Y)
LMLE_k3 = 0
for i in range(Y.shape[0]):
    LMLE_k3 += np.log(norm.pdf(Y[i],loc=np.dot(np.squeeze(wmle_k3),pdata_k3[i]),scale = np.sqrt(1/beta)))
print("MLE for K=3",LMLE_k3)
k=4 ###k=4
wmle_k4 = np.matmul(np.matmul(np.linalg.inv(np.matmul(pdata_k4.T,pdata_k4)),pdata_k4.T),Y)
LMLE_k4 = 0
for i in range(Y.shape[0]):
    LMLE_k4 += np.log(norm.pdf(Y[i],loc=np.dot(np.squeeze(wmle_k4),pdata_k4[i]),scale = np.sqrt(1/beta)))
print("MLE for K=4",LMLE_k4)
#print(np.log(norm.pdf(Y[0],loc=np.dot(np.squeeze(wmle_k1),pdata_k1[0]),scale = np.sqrt(1/beta))))
```
